```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\Imran\Documents\women_in_news\headlines.xlsx")
```

```python
df.head
```

```python
df['headline'] = df['headline'].str.lower()
```

```python
contain_values = df[df['headline'].str.contains('women|woman|girls|girl|wife|daughter|sister|female|lady|mother|actress| her | she | mrs | niece | maid | servant|shcoolgirl')]
print (contain_values)
```

```python
filtered_df = contain_values[~contain_values['headline'].str.contains('mother language')]
print (filtered_df)
```

```python
headline_list = filtered_df["headline"].tolist()
print(headline_list)
```

```python
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize

joint_words = ' '.join(headline_list)
tokenized_words = word_tokenize(joint_words)

print(tokenized_words)
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
print(len(stop_words))
print (stop_words)
```

```python
#adding my own stop words
stop_words.update(['.', ',', '"', "'", ':', ';', '!', '(', ')', '{', '}', '``', '``', '‘', '[', ']', '•', 
                   '’', 'fy', '?','▪', 'taka', 'fiscal', 'year', 'years', 'per', 'cent', 'order', 'percent', 
                   'would', 'mr', 'mr.', 'also', 'already', 'rate', 'sector', 'lakh','project', 'new', 'time', 
                   'propose', 'proposed', 'thousand', "'s", "dy", 'two', '2', '3', 'schoolgirl',
                   'taken', 'total', 'step', 'total''total', '%', 'speaker', 'tk', 'crore', 'sir', "''", 'women', 
                   'woman', 'girls','girl','wife','daughter','sister','female','lady','mother','actress', 'housewife',
                   'her', 'she', 'mrs', 'niece', 'maid','servant'])
print(len(stop_words))
print(stop_words)
```

```python
fltrd_words = [word for word in tokenized_words if not word in stop_words]
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lem_tokens = [lemmatizer.lemmatize(word) for word in fltrd_words]
from nltk import FreqDist
fltrd_words_dist= FreqDist(lem_tokens)
import matplotlib.pyplot as plt

plt.xticks(fontsize=13)
fltrd_words_dist.plot(25,title='Freq. Dist. of 15 most common words')
fltrd_words_dist.most_common(25)
```
